```python
import os
import fastf1 as ff1

# 1. 'cache' 폴더가 없으면 생성합니다.
if not os.path.exists('cache'):
    os.makedirs('cache')

# 2. 이제 캐시를 활성화합니다.
ff1.Cache.enable_cache('cache')
```

```python
# 1. 세션 정의
session = ff1.get_session(2025, 'Abu Dhabi', 'Q')
```

```python
# 2. 데이터 로딩 (여기서 텔레메트리까지 한꺼번에 불러옵니다)
# 이전의 load_laps() 대신 아래 방식을 사용하세요.
session.load(telemetry=True)

# 3. 랩 데이터 접근
# 이제 session.laps를 통해 모든 랩 정보를 가져올 수 있습니다.
laps = session.laps

# 결과 확인: 가장 빠른 랩 출력해보기
fastest_lap = laps.pick_fastest()
print(f"가장 빠른 드라이버: {fastest_lap['Driver']}")
print(f"기록: {fastest_lap['LapTime']}")
```
